```python
import re
from unidecode import unidecode

def clean_text(text, mode='Keywords'):
    if mode=='Keywords':
        # keyword clean
        result = re.sub(r'\s*-\s*', '-', text)
        result = re.split(r'[^\w-]',result)
    else:
        # Title and name clean
        result=re.split(r'[^\w]',text)
    result=[s.lower() for s in result if s!='']
    if mode!=1:
        result=[unidecode(word) for word in result]
    return result

clean_text("   web'\(trang web - 1090 !); very  \n good",mode='Keywords')
# clean_text('Duc - Trong Le',mode='Name')
# clean_text('Kiến trúc nhiều tầng cho phát hiện và ngăn chặn trang web lừa đảo',mode='Title')
```

```python
name_variation=[]
variation_origin={}
for idx,row in teacher_uet.iterrows():
    variation_list=row['Name variation'].split(',')
    for variation in variation_list:
        variation_origin[variation]=row['Tên']
    name_variation.extend(variation_list)

clean_title_set=[]
clean_raw_title={}
raw_title_set=paper_uet['Title'].to_list()+thesis_uet['Title'].to_list()+thesis_concate['Title'].to_list()
for raw_title in raw_title_set:
    clean_title=' '.join(clean_text(raw_title, mode='Title'))
    clean_raw_title[clean_title]=raw_title
    clean_title_set.append(clean_title)
```

```python
from thefuzz import fuzz,process

def search_name(name):
    name=' '.join(clean_text(name, mode='Name'))
    candidate=process.extractBests(name, name_variation,
                     limit=10,
                     scorer=fuzz.token_sort_ratio)

    candidate=[i[0] for i in candidate]
    candidate=process.extractBests(name, candidate,
                     limit=5,
                     scorer=fuzz.WRatio)

    max_score=candidate[0][1]
    candidate=[variation_origin[i[0]] for i in candidate if i[1]==max_score]
    return candidate,max_score

# search_name('Hà Quang Thụy')
# search_name('haqt')
# search_name('thuy tran truong')
search_name('tthuyt')
```

```python
from thefuzz import fuzz,process

def search_title(title):
    title=' '.join(clean_text(title, mode='Title'))
    candidate=process.extractBests(title, clean_title_set,
                     limit=1,
                     scorer=fuzz.WRatio)
    final=[i[0] for i in candidate if i[1]==candidate[0][1]]
    final=[clean_raw_title[i] for i in final]
    return final,candidate[0][1]

search_title('Machine learning-based extraction of semantic relations from biomedical literature')
```

```python
from llama_index.core.retrievers import BaseRetriever
from llama_index.core.postprocessor import SimilarityPostprocessor
# from llama_index.postprocessor.cohere_rerank import CohereRerank
from llama_index.core.postprocessor import SentenceTransformerRerank
from llama_index.core.postprocessor import LongContextReorder
from llama_index.core import QueryBundle


class HybridRetriever(BaseRetriever):
    def __init__(self, vector_retriever, vector_post_process,bm25_retriever, bm25_post_process,final_post_process):
        self.vector_retriever = vector_retriever
        self.vector_post_process = vector_post_process
        self.bm25_retriever = bm25_retriever
        self.bm25_post_process = bm25_post_process
        self.final_post_process = final_post_process
        super().__init__()

    def _retrieve(self, query, **kwargs):
        bm25_nodes = self.bm25_post_process.postprocess_nodes(self.bm25_retriever.retrieve(query, **kwargs))
        vector_nodes = self.vector_post_process.postprocess_nodes(self.vector_retriever.retrieve('query: '+query, **kwargs))

        all_nodes = []
        node_ids = set()
        for n in bm25_nodes + vector_nodes:
            if n.node.node_id not in node_ids:
                all_nodes.append(n)
                node_ids.add(n.node.node_id)

        for post_process in self.final_post_process:
            all_nodes = post_process.postprocess_nodes(all_nodes,query_bundle=QueryBundle('query: '+query))

        return all_nodes


table_final_post_process = [
    # CohereRerank(model="rerank-english-v2.0", api_key="Q7DQqKxPaBsdL44ZF77mO07SkUtcpmMRihrIq8RY",top_n=5),
    SentenceTransformerRerank(model="BAAI/bge-reranker-v2-m3",top_n=40),
    LongContextReorder()
]
hybrid_paper_uet = HybridRetriever(vector_paper_uet.as_retriever(similarity_top_k=100), SimilarityPostprocessor(similarity_cutoff=0.5), bm25_paper_uet,SimilarityPostprocessor(similarity_cutoff=0.1),table_final_post_process)
hybrid_thesis_uet = HybridRetriever(vector_thesis_uet.as_retriever(similarity_top_k=100), SimilarityPostprocessor(similarity_cutoff=0.5), bm25_thesis_uet, SimilarityPostprocessor(similarity_cutoff=0.1),table_final_post_process)
hybrid_thesis_concate = HybridRetriever(vector_thesis_concate.as_retriever(similarity_top_k=100), SimilarityPostprocessor(similarity_cutoff=0.5), bm25_thesis_concate, SimilarityPostprocessor(similarity_cutoff=0.1),table_final_post_process)

prose_final_post_process = [
    # CohereRerank(model="rerank-english-v2.0", api_key="Q7DQqKxPaBsdL44ZF77mO07SkUtcpmMRihrIq8RY",top_n=5),
    SentenceTransformerRerank(model="BAAI/bge-reranker-v2-m3",top_n=3),
    LongContextReorder()
]
hybrid_thesis_perform = HybridRetriever(vector_thesis_perform.as_retriever(similarity_top_k=20), SimilarityPostprocessor(similarity_cutoff=0.1), bm25_thesis_perform, SimilarityPostprocessor(similarity_cutoff=0.1),prose_final_post_process)
```

```python
from langchain.agents import tool

@tool
def get_personal_info():
    """Get user's personal information. Use when user asking information about themself"""
    context='Here is user information: \n'
    not_result=['Name','School','Total Credit']
    for col in not_result:
        context=context+col+': '+str(personal_info.loc[0,col])+'\n'

    context=context+'Learning result:\n'
    for col in personal_info.columns:
        if col in not_result:
            continue
        else:
            context=context+'- '+col+': '+str(personal_info.loc[0,col])+'\n'

    context=context+'\n\n'+'*'*20+'\n\n'
    return context

print(get_personal_info.run({}))
```

```python
from pydantic.v1 import BaseModel,Field

class Intent(BaseModel):
    """Tag the piece of text with particular infor."""
    query: str = Field(..., description="information required by user")

@tool(args_schema=Intent)
def get_thesis_perform(query):
    """Get general information of doing thesis: process, qualification, etc"""
    retrieved=hybrid_thesis_perform._retrieve(query)
    context=''
    for node in retrieved:
        context=context+'\n'+node.text
    return context

# print(get_thesis_perform.run({'query': 'Ngày nộp khóa luận'}))
```

```python
class Object_Name(BaseModel):
    """Tag the piece of text with particular infor."""
    target: str = Field(..., description="Title of a work or name of a person")

@tool(args_schema=Object_Name)
def get_teacher_info(target):
    """Use it to get information about a specific teacher. Use it only when you are sure about the query intention."""

    name,score=search_name(target)
    context=''
    info=teacher_uet[teacher_uet['Tên'].isin(name)]
    interest=['Tên', 'Bộ môn', 'Chức danh', 'Hướng nghiên cứu', 'Chủ đề khóa luận', 'Liên hệ', 'HomePage', 'Link', 'Number of paper']
    for idx,row in info.iterrows():
        tmp=''
        for col in interest:
            if row[col]!='':
                tmp=tmp+col+": "+str(row[col])+'\n'
        context=context+tmp+'*'*20+'\n\n'

    return context

@tool(args_schema=Object_Name)
def get_single_work_info(target: str):
    """Use to get information about a specific paper or thesis given its title. Use it only when you are sure about the query intention."""
    clean_title=' '.join(clean_text(target,mode='Title'))
    candidate=search_title(clean_title)
    clean_title=candidate[0][0]
    # if candidate[1]<60:
        # return 'There no work name '+ target +'\n'

    info=paper_uet[paper_uet['Title']==clean_title]
    if len(info)==0:
        info=thesis_uet[thesis_uet['Title']==clean_title]
    if len(info)==0:
        info=thesis_concate[thesis_concate['Title']==clean_title]
    if len(info)==0:
        return 'There no work name '+ target +'\n'

    interest=['Title','Abstract','Advisor','Link','Description','Authors']
    context=''
    for col in interest:
        if col in info.columns:
            context=context+col+': '+info.iloc[0][col]+'\n\n'

    context=context+'*'*20+'\n'
    return context

# print(get_teacher_info.run({'target': 'lsvinh'}))
# print(get_single_work_info.run({'target':'Machine learning-based extraction of semantic relations from biomedical literature'}))
```

```python
import random
import numpy as np

def sample_node(node_list,k=5, mode=1):
    if len(node_list)==0:
        return []
    if mode==1:
        score=[i.score for i in node_list]
    else:
        score=np.array([1.0 for i in node_list])

    num_samples = min(k, len(node_list))
    sampled_keys = np.random.choice(range(len(node_list)), size=num_samples, replace=False, p=score/sum(score))
    return [node_list[idx] for idx in sorted(sampled_keys)]

def filt_doc(doc_store,field,separator,name_variation):
    document=[]
    for node in doc_store:
        for key, value in node.metadata.items():
            if key!=field:
                continue
            if value=='':
                  continue
            name=[' '.join(clean_text(i,mode='Name')) for i in value.split(separator)]

            exist_name=[]
            for i in name:
                score=process.extractBests(i, name_variation, limit=1,scorer=fuzz.token_sort_ratio)[0][1]
                if score>=90:
                    exist_name.append(i)
                    break

            if len(exist_name)!=0:
                document.append(node)

    return document

def retrieve_document(keyword,teacher, retriever):

    field=''
    separator=''

    if retriever==hybrid_paper_uet:
        field='Authors'
        separator=','
    else:
        field='Advisor'
        separator='\n'

    keyword=', '.join(clean_text(keyword,mode='Keywords'))

    if keyword!='' and teacher!='':
        name,score=search_name(teacher)
        name_variation=[]
        for idx,row in teacher_uet[teacher_uet['Tên'].isin(name)].iterrows():
            name_variation.extend(row['Name variation'].split(','))

        retrieved=retriever._retrieve(keyword)
        document=document=filt_doc(retrieved,field,separator,name_variation)

        return sample_node(document)
        # ****************************************************************

    elif keyword=='' and teacher!='':
        name,score=search_name(teacher)
        name_variation=[]

        for idx,row in teacher_uet[teacher_uet['Tên'].isin(name)].iterrows():
            name_variation.extend(row['Name variation'].split(','))

        doc_store=''
        if retriever==hybrid_paper_uet:
            doc_store=paper_uet_nodes
        elif retriever==hybrid_thesis_uet:
            doc_store=thesis_uet_nodes
        elif retriever==hybrid_thesis_concate:
            doc_store=thesis_concate_nodes

        document=filt_doc(doc_store,field,separator,name_variation)

        return sample_node(document,mode=2)
# ****************************************************************

    elif keyword!='' and teacher=='':
        retrieved=retriever._retrieve(keyword)
        return sample_node(retrieved)

    else:
        return []

# retrieved=retrieve_document('','Lê Sỹ Vinh', hybrid_paper_uet)
# retrieved=retrieve_document('machine leaning','thuy hg', hybrid_thesis_uet)
# retrieved=retrieve_document('web','', hybrid_thesis_concate)
retrieved=retrieve_document('','tran trường thủy',hybrid_paper_uet)
for node in retrieved:
    print(node.metadata)
  # display_source_node(node)
# search_name('tran trường thủy')
# process.extractBests('thuytt', ['tran truong thuy', 'tt thuy', 'tt tran'], limit=1,scorer=fuzz.ratio)
# process.extractBests('the trung tran', ['tran truong thuy', 'tt thuy', 'tt tran'], limit=1,scorer=fuzz.WRatio)
# the trung tran,
# thuy trang tran
# tran thu trang
```

```python
class Tagging(BaseModel):
    """tag the piece of text with particular infor."""
    keyword: str = Field(...,description="Phrases that best express the document subject and type, separated by commas.")
    teacher: str = Field(...,description="A single name is desired or suggested as the as the author or advisor of the query. If not exist, let it a empty string")

def concate_node(node_list):
    context=''
    for node in node_list:
        tmp=node.text
        for key,value in node.metadata.items():
            if value=='':
                  continue
            tmp=tmp+'\n'+key+': '+value
        context=context+tmp+'\n\n'

    return context

@tool(args_schema=Tagging)
def contextualize_paper_uet_info(keyword: str,teacher: str) -> str:
    """Retrieve papers of uet"""
    retrieved=retrieve_document(keyword,teacher,hybrid_paper_uet)
    if len(retrieved)==0:
        context='There no paper related to '+  keyword
        if teacher!='':
            context=context+' and may contributed by '+teacher
        return context

    context="Here is the list of paper that related to "+keyword
    if teacher!='':
        context=context+' and may contributed by '+teacher
    context=context+':\n\n'

    context=context+concate_node(retrieved)

    context=context+'The information above may contain information not related to the user query. Think carefully, then use the useful information to answer the user query.'+'\n\n'
    context=context+'*'*20
    return context

@tool(args_schema=Tagging)
def contextualize_thesis_uet_info(keyword: str,teacher: str) -> str:
    """Retrieve uet thesis"""
    retrieved=retrieve_document(keyword,teacher,hybrid_thesis_uet)

    if len(retrieved)==0:
        context='There no thesis related to '+  keyword
        if teacher!='':
            context=context+' and may advised by '+teacher
        return context

    context="Here is the list of UET thesis that related to "+keyword
    if teacher!='':
        context=context+' and may advised by '+teacher
    context=context+':\n\n'

    context=context+concate_node(retrieved)

    context=context+'The information above may contain information not related to the user query. Think carefully, then use the useful information to answer the user query.'+'\n\n'
    context=context+'*'*20
    return context

@tool(args_schema=Tagging)
def contextualize_thesis_concate_info(keyword: str,teacher: str) -> str:
    """Retrieve thesis that not belong to uet"""
    retrieved=retrieve_document(keyword,teacher, hybrid_thesis_concate)
    if len(retrieved)==0:
        context='There no thesis related to '+  keyword
        if teacher!='':
            context=context+' and may advised by '+teacher
        return context

    context="Here is the list of thesis that not belong to UET and may related to "+keyword
    context=context+':\n\n'

    context=context+concate_node(retrieved)

    context=context+'The information above may contain information not related to the user query. Think carefully, then use the useful information to answer the user query.'+'\n\n'
    context=context+'*'*20
    return context

@tool(args_schema=Tagging)
def contextualize_thesis_general(keyword: str,teacher: str) -> str:
    """Retrieve thesis when the query is not specific the thesis should belong to uet or not."""

    retrieved=retrieve_document(keyword,teacher, hybrid_thesis_uet)

    if teacher=='':
        retrieved=retrieved+retrieve_document(keyword,teacher, hybrid_thesis_concate)

    if len(retrieved)==0:
        context='There no thesis related to '+  keyword
        if teacher!='':
            context=context+' and may advised by '+teacher
        return context

    context="Here is the list of thesis may related to "+keyword
    if teacher!='':
        context=context+' and may advised by '+teacher
    context=context+':\n\n'

    context=context+concate_node(retrieved)

    context=context+'The information above may contain information not related to the user query. Think carefully, then use the useful information to answer the user query.'+'\n\n'
    context=context+'*'*20
    return context

# print(contextualize_paper_uet_info.run({'keyword': 'machine learning', 'teacher': 'Lê Sỹ Vinh'}))
# print(contextualize_thesis_uet_info.run({'keyword': 'web', 'teacher': 'thuyqh'}))
# print(contextualize_thesis_uet_info.run({'keyword': 'học máy', 'teacher': ''}))

# print(contextualize_thesis_concate_info.run({'keyword' : 'machine learning', 'teacher' : ''}))
# print(contextualize_thesis_general.run({'keyword' : 'machine learning', 'teacher' : ''}))
```

```python
@tool(args_schema=Tagging)
def suggest_teacher(keyword: str,teacher: str):
    """Seaching for research direction of reacher. Use it only when the user asks for a suggested thesis advisor or asks for a teacher with research related to some subject."""

    teacher_research=retrieve_document(keyword, '', hybrid_paper_uet)
    teacher_thesis=retrieve_document(keyword, '', hybrid_thesis_uet)
    candidate=[]
    context='Here are some works that are related to: '+keyword+'\n\n'

    for node in teacher_research:
        for key, value in node.metadata.items():
            if key!='Authors':
                  continue
            name=[' '.join(clean_text(i,mode=1)) for i in value.split(',')]
            candidate.extend(name)

    context=context+concate_node(teacher_research)+'\n'

    for node in teacher_thesis:
        for key, value in node.metadata.items():
            if value!='Advisor':
                  continue
            name=[' '.join(clean_text(i,mode=1)) for i in value.split('\n')]
            candidate.extend(name)

    context=context+concate_node(teacher_thesis)+'\n\n'

    context=context+'Here is a name glosarry:\n\n'
    suggest_teacher=[]
    for i in candidate:
        tmp,score=search_name(i)
        if score>80:
            suggest_teacher.append(tmp[0])
    suggest_teacher=list(set(suggest_teacher))
    teacher_data = teacher_uet[teacher_uet['Tên'].isin(suggest_teacher)]

    for idx,row in teacher_data.iterrows():
        context=context+'Name: '+row['Tên']+'\n'+'Some name variation example: '+row['Name variation']+'\n\n'

    context=context+'Use the infomration above to suggest some teacher that in name grossary'+'\n\n'
    context=context+'*'*20+'\n\n'

    return context

# print(suggest_teacher.run({'keyword':'web', 'teacher': ''}))
```

```python
import pandas as pd
from deepeval import evaluate
from deepeval import assert_test
from deepeval.metrics import AnswerRelevancyMetric,FaithfulnessMetric,ContextualRelevancyMetric
from deepeval.test_case import LLMTestCase

answer_relevancy_metric = AnswerRelevancyMetric(model='gpt-3.5-turbo',
                                                threshold=0.07,
                                                async_mode=False,
                                                include_reason=True)
faithfulness_metric =  FaithfulnessMetric(threshold=0.7,
                                          model='gpt-3.5-turbo',
                                          async_mode=False,
                                          include_reason=True)
contextual_relevancy_metric = ContextualRelevancyMetric(threshold=0.7,
                                                        model="gpt-3.5-turbo",
                                                        include_reason=True)
def eval_llm_response(query,context,response):
    test_case = LLMTestCase(input=query,
                            actual_output=response,
                            # expected_output
                            retrieval_context=context)
    res=evaluate([test_case],
                 [answer_relevancy_metric,faithfulness_metric,contextual_relevancy_metric],
                 run_async=False,
                 show_indicator=False)
    df=pd.DataFrame()

    for idx,metric in enumerate(res[0].metrics):
        df.loc[idx,'Metric']=metric.__name__
        df.loc[idx,'Evaluation_cost']=metric.evaluation_cost
        df.loc[idx,'Score']=metric.score
        df.loc[idx,'Reason']=metric.reason

    # df.index=[i.__name__ for i in res[0].metrics]
    return df

eval_llm_response("What if these shoes don't fit?",
     ["All customers are eligible for a 30 day full refund at no extra costs."],
     "We offer a 30-day full refund at no extra costs.")
```

```python
empty_df=pd.DataFrame({'evaluation_cost': ['No information'],
                       'score': ['No information'],
                       'reason': ['No information']})

def gradio_eval_function(chat_history,context):
    if len(chat_history)==0 or len(context[0][0])==0:
        return empty_df

    response=chat_history[-1][1]

    chat_len=len(chat_history)
    chat_len=max(1,2)
    chat_context=''
    for chat in chat_history[-(chat_len-1):]:
        chat_context=chat_context+chat[0]+'\n\n'+chat[1]+'\n\n'
    chat_context=chat_history[-1][0]
    info_context=context[0][0]
    return eval_llm_response(chat_context,[info_context],response)

# gradio_eval_function([['i love cat','That sound grate']],
#                       [['I feed cat everyday',]])
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain.prompts import MessagesPlaceholder

prompt=ChatPromptTemplate.from_messages([
    ('system', '{instruction}'),
    MessagesPlaceholder(variable_name='chat_history'),
    ('user','{input}'),
    MessagesPlaceholder(variable_name='agent_scratchpad')

])
```

```python
from queue import SimpleQueue
from typing import Any, Dict, List, Union
from langchain.callbacks.base import BaseCallbackHandler
from langchain.schema import LLMResult

q = SimpleQueue()
job_done = object()
class StreamingGradioCallbackHandler(BaseCallbackHandler):
    """Callback handler - works with LLMs that support streaming."""

    def __init__(self, q: SimpleQueue):
        self.q = q

    def on_llm_start(
        self, serialized: Dict[str, Any], prompts: List[str], **kwargs: Any
    ) -> None:
        """Run when LLM starts running."""
        while not self.q.empty():
            try:
                self.q.get(block=False)
            except SimpleQueue.empty:
                continue

    def on_llm_new_token(self, token: str, **kwargs: Any) -> None:
        """Run on new LLM token. Only available when streaming is enabled."""
        self.q.put(token)

    def on_llm_end(self, response: LLMResult, **kwargs: Any) -> None:
        """Run when LLM ends running."""
        self.q.put(job_done)

    def on_llm_error(
        self, error: Union[Exception, KeyboardInterrupt], **kwargs: Any
    ) -> None:
        """Run when LLM errors."""
        self.q.put(job_done)
```

```python
import time
import threading
from langchain_openai import ChatOpenAI,OpenAI
from langchain.agents import AgentExecutor
from langchain.memory import ConversationBufferWindowMemory
from langchain.schema.runnable import RunnablePassthrough
from langchain.agents.format_scratchpad import format_to_openai_functions
from langchain.agents.output_parsers import OpenAIFunctionsAgentOutputParser
from langchain.callbacks.streaming_stdout_final_only import StreamingStdOutCallbackHandler

memory=ConversationBufferWindowMemory(k=5,return_messages=True,memory_key='chat_history',input_key='input', output_key="output")

def base_generate(message,chat_history,instruction,model,temperature,max_tokens):
    if len(chat_history)==0:
        memory.clear()

    llm = ChatOpenAI(model=model,
                     temperature=temperature,
                     max_tokens=max_tokens,
                     streaming=True,
                     callbacks=[StreamingGradioCallbackHandler(q),
                                StreamingStdOutCallbackHandler()])

    model_with_function=llm.bind(functions=funtions)
    agent_chain=RunnablePassthrough.assign(
        instruction = lambda x : instruction,
        agent_scratchpad= lambda x: format_to_openai_functions(x['intermediate_steps'])
    ) | prompt | model_with_function | OpenAIFunctionsAgentOutputParser()


    agent_executor=AgentExecutor(agent=agent_chain,tools=tools,memory=memory,
                                 chain_type='map_reduce',
                                 return_intermediate_steps=True,verbose=False,
                                 early_stopping_method='generate',
                                #  handle_parsing_errors=True, max_iterations=1
                                 )

    response=agent_executor.invoke({"input": message})
    context=[i[1] for i in response['intermediate_steps']]
    context=[[i,None] for i in context]
    return context

def generate(message, chat_history, instruction='You are a helpful assistant', model='gpt-3.5-turbo', temperature=0.7, max_tokens=156):
    def generate_in_thread(message, chat_history, instruction, model, temperature, max_tokens):
        global output
        output = base_generate(message, chat_history, instruction, model, temperature, max_tokens)

    thread = threading.Thread(target=generate_in_thread, args=(message, chat_history, instruction, model, temperature, max_tokens,))
    thread.start()

    chat_history = chat_history + [[message, None]]
    yield '', chat_history,[]
    chat_history[-1][1] = ''

    while True:
        time.sleep(0.08)
        next_token = q.get(block=True)
        if next_token is job_done:
            break
        chat_history[-1][1] = chat_history[-1][1] + next_token
        yield '', chat_history,[]
    thread.join()
    yield '',chat_history,output

chat_his=[
    ['Hi','Hello! How can I assist you today?'],
    ['Can you tell me the meaning of life',"The meaning of life is a philosophical question that has been debated for centuries and varies depending on individual beliefs and perspectives. Some people find meaning in religion, relationships, personal growth, helping others, or pursuing their passions. It's a deeply personal and subjective question that each person may answer differently based on their own experiences and values."],
    # ['Can you introduce me to piano','']
]
```
